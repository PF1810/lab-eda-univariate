```python
#1.
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import skew, kurtosis

path_amazon_uk = "C:\\Users\\pedro\\Documents\\ironhack\\Course\\WEEK 5\\LAB 1\\lab-eda-univariate\\amz_uk_price_prediction_dataset.csv"

df = pd.read_csv(path_amazon_uk)

df.head()
```

```python
#2.
# Frequency table for the product category
category_frequency = df['category'].value_counts()


# Top 5 most listed product categories
top_5_categories = category_frequency.head(5)
top_5_categories
```

```python
#3.
# Bar chart for product categories
plt.figure(figsize=(12, 6))
category_frequency.plot(kind='bar', color='skyblue')
plt.title('Distribution of Products Across Different Categories')
plt.xlabel('Category')
plt.ylabel('Number of Listings')
plt.show()
```

```python
#4.
# Pie chart for top 5 categories
plt.figure(figsize=(8, 8))
top_5_categories.plot(kind='pie', autopct='%1.1f%%')
plt.title('Proportion of Top 5 Product Categories')
plt.ylabel('')
plt.show()

```

```python
#4
# Horizontal bar chart for top 5 categories
plt.figure(figsize=(10, 6))
top_5_categories.plot(kind='barh', color='skyblue')
plt.title('Proportion of Top 5 Product Categories')
plt.xlabel('Number of Listings')
plt.ylabel('Product Category')
plt.show()
```

```python
#5.
# Mean, median, and mode for the price of products
mean_price = df['price'].mean()
median_price = df['price'].median()
mode_price = df['price'].mode()[0]

mean_price, median_price, mode_price
```

```python
#6.
# Variance, standard deviation, range, and interquartile range for product price
variance_price = df['price'].var()
std_dev_price = df['price'].std()
range_price = df['price'].max() - df['price'].min()
iqr_price = df['price'].quantile(0.75) - df['price'].quantile(0.25)

variance_price, std_dev_price, range_price, iqr_price
```

```python
#7.
# Histogram for product prices
plt.figure(figsize=(12, 6))
plt.hist(df['price'], bins=50, edgecolor='black')
plt.title('Distribution of Product Prices')
plt.xlabel('Price')
plt.ylabel('Number of Products')
plt.show()
```

```python
#8.
# Box plot for product prices
plt.figure(figsize=(12, 6))
plt.boxplot(df['price'], vert=False)
plt.title('Box Plot of Product Prices')
plt.xlabel('Price')
plt.show()
```

```python
#9.
# Mean, median, and mode for the rating of products
mean_rating = df['stars'].mean()
median_rating = df['stars'].median()
mode_rating = df['stars'].mode()[0]

mean_rating, median_rating, mode_rating
```

```python
#10.
# Variance, standard deviation, and interquartile range for product rating
variance_rating = df['stars'].var()
std_dev_rating = df['stars'].std()
iqr_rating = df['stars'].quantile(0.75) - df['stars'].quantile(0.25)

variance_rating, std_dev_rating, iqr_rating
```

```python
#11.
# Skewness and kurtosis for product rating
skewness_rating = skew(df['stars'])
kurtosis_rating = kurtosis(df['stars'])

skewness_rating, kurtosis_rating
```

```python
#12.
# Histogram for product ratings
plt.figure(figsize=(12, 6))
plt.hist(df['stars'], bins=10, edgecolor='black')
plt.title('Distribution of Product Ratings')
plt.xlabel('Rating')
plt.ylabel('Number of Products')
plt.show()
```
